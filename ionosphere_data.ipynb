```python
import tensorflow as tf
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ion_data=pd.read_csv('./ionosphere_data.csv')
```

```python
ion_data.head()
```

```python
ion_data=ion_data.sample(frac=1)
```

```python
ion_data.drop('feature2',
 axis='columns', inplace=True)
```

```python
ion_data.head()
```

```python
ion_data['label']=ion_data['label'].map({'g':1,"b":0})
```

```python
ion_data.head()
```

```python
data=ion_data.iloc[:,:33]
labels=ion_data.iloc[:,33:]
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(data,labels, test_size=0.40)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
mean=x_train.mean()
std=x_train.std()
x_train -= mean
x_train /=std
x_test -= mean
x_test /=std
```

```python
x_test.head()
```

```python
x_train.head(3)
```

```python
x_test.head(3)
```

```python
train=x_train
```

```python
test=x_test
```

```python
train.head()
```

```python
train_label=np.asarray(y_train)
test_label=np.asarray(y_test)
```

```python
# now we have the following
#train
#test
#train_label
#test_label
```

```python
train.shape
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
mymodel=models.Sequential()
mymodel.add(layers.Dense(16, activation='relu', input_shape=(train.shape[1],)))
mymodel.add(layers.Dense(12, activation='relu'))
mymodel.add(layers.Dense(1, activation='sigmoid'))
```

```python
mymodel.compile(optimizer='rmsprop',
 loss='binary_crossentropy',
 metrics=['accuracy'])
```

```python
mymodel.fit(train,train_label, epochs=100,batch_size=1)
```

```python
results = mymodel.evaluate(test, test_label)
```

```python
pred=mymodel.predict(test)
pred[0:10]
```

```python
train_label[0:10]
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import regularizers
mymodel=models.Sequential()
mymodel.add(layers.Dense(16,kernel_regularizer=regularizers.l2(0.001),
 activation='relu', input_shape=(train.shape[1],)))
mymodel.add(layers.Dense(12,kernel_regularizer=regularizers.l2(0.001),
 activation='relu'))
mymodel.add(layers.Dense(1, activation='sigmoid'))
```

```python
mymodel.compile(optimizer='rmsprop',
 loss='binary_crossentropy',
 metrics=['accuracy'])
```

```python
mymodel.fit(train,train_label, epochs=100,batch_size=1, verbose=0)
```

```python
results = mymodel.evaluate(test, test_label)
results
```

```python
pred=mymodel.predict(test)
pred[0:10]
```
